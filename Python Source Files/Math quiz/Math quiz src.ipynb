```python
#Tkinter GUI math quiz.
import time
import pandas as pd
import tkinter as tk
from tkinter import messagebox

df = pd.read_csv("Math quiz question bank.csv")
df = df.dropna()
score = 0
question_line = 0
question_number = 1

def question_switch():
    global question_line, question, answer, question_number
    question_line += 1
    question_number += 1

    if question_line >= len(df):
        return None,None
    
    question = df.iloc[question_line,0]
    answer = df.iloc[question_line,1]

    if pd.isna(question) or pd.isna(answer):
        return None, None
    return question, answer


def submit():
    global score, answer
    answer_given = question_entry.get().upper()
    if answer_given != answer:
        messagebox.showinfo("Whoops!",f"Wrong answer, the answer was {answer}")
    else:
        messagebox.showinfo("Hooray!", "You got it right!")
        score += 1
    result = question_switch()
    if result == (None, None):
        question_label.config(text = f"Quiz is finished, your score is {score} out of {question_line}({(score/(question_line)) * 100}%)")
        submit_button.destroy()
        return
    question_label.config(text = f"{question_number}) {question}")
    question_entry.delete(0, tk.END)

question = df.iloc[0,0]
answer = df.iloc[0,1]

root = tk.Tk()
root.title("Cerebus math quiz (with Tkinter)")
root.geometry("800x800")

messagebox.showinfo("Get ready!", 'Your quiz will be loaded in 3 seconds.')
time.sleep(3)
messagebox.showinfo("Start!", "Your quiz has started.")

question_label = tk.Label(root, text = f"{question_number}) {question}")
question_label.pack()
question_entry = tk.Entry()
question_entry.pack()

submit_button = tk.Button(root, text = "Submit answer", command = submit)
submit_button.pack()
root.mainloop()

```
